```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/drive/MyDrive/banking_data.csv')
```

```python
df
```

```python
df=df.drop('ID',axis=1)
```

```python
df
```

```python
x = df[['Salary (INR/month)', 'Credit Score', 'Loan Amount (INR)']]
y = df['Loan Approved']
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x,y)
```

```python
test_point=[[65000,805,800000]]
```

```python
prediction=knn.predict(test_point)
print("Predicted Class:",prediction[0])
```

```python
colors=df['Loan Approved'].map({'Yes':'green','No':'red'})
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Salary (INR/month)'],df['Credit Score'],c=colors,label='Test Point',s=100,edgecolors='black')
plt.xlabel('Salary (INR/month)')
plt.ylabel('Credit Score')
plt.title('K-Nearest Neighbors Classification')
plt.legend()
plt.show()
```
